# Exploratory Data Analysis

### Gender

```python
ax = sns.countplot(x ="Gender", data=df, hue="Gender")
sns.set(rc={"figure.figsize":(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount
sales_gen = df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
ax = sns.barplot(x="Gender", y="Amount", data=sales_gen, hue="Gender")
sns.set(rc={"figure.figsize":(8,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Amount spent by Age Group:
Spent_by_Grp = df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sns.barplot(x="Age Group", y="Amount", data = Spent_by_Grp, hue="Age Group")

```

### States

```python
ax = sns.countplot(x="Age Group", data=df, hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Spent_by_States = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={"figure.figsize":(18,8)})
sns.barplot(x="State", y = "Amount", data=Spent_by_States, hue="State")
```

### Occupation

```python
by_occuptaion = df.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
sns.barplot(x="Occupation", y="Amount", data=by_occuptaion, hue="Occupation")
```

### Product Category

```python
by_product  = df.groupby(["Product"], as_index=False)["Amount"].sum().sort_values(by="Amount").head(10)
sns.barplot(x = "Product", y="Amount", data=by_product, hue="Product")
```
